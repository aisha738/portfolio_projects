```python
#function to get top n zipcodes
def top_unique_zips(df, n):
    # Group the df by unique zip codes and calculate the mean price for each zip code
    zip_price_mean = df.groupby('zipcode')['price'].mean().reset_index()
    
    # Sort the zip codes by mean price
    sorted_zipcodes = zip_price_mean.sort_values(by='price', ascending=False)
    
    # Select the top N zip codes
    top_n_zips = sorted_zipcodes.head(n)['zipcode']
    
    return top_n_zips


#function to get bottom n zipcodes
def bottom_unique_zips(df, n):
    # Group the df by unique zip codes and calculate the mean price for each zip code
    zip_price_mean = df.groupby('zipcode')['price'].mean().reset_index()
    
    # Sort the zip codes by mean price
    sorted_zipcodes = zip_price_mean.sort_values(by='price', ascending=False)
    
    # Select the bottom N zip codes
    bottom_n_zips = sorted_zipcodes.tail(n)['zipcode']
    
    return bottom_n_zips
```

```python
# Creating a function that applies the bedroom threshold criteria to form 2 groups
# and then another function that applies the first function to create a new column 'prop_type' 

def prop_group(num_bedrooms):
    if num_bedrooms >= 2.5:
        return 'Single-Family Home'
    else:
        return 'Condo'

def segment_prop_types(df):
    df['prop_type'] = df['bedrooms'].apply(prop_group)
    return df
    
```

- I used bedrooms for the segmentation, because it is common to have either equal number of bedrooms and bathrooms, or more bedrooms than bathrooms.


- This helps me to prevent overlap, eg. a case where the property can fall under both 'Apartment' and 'single-family home' based on the bedroom and bathroom threshold criteria.

```python
# Grouping dataset by property type 
grp = df.groupby('prop_type')
grp.first()
```

```python
# subsetting the dataframe
condo_df = df[df['prop_type'] == 'Condo']
sing_fam_df = df[df['prop_type'] == 'Single-Family Home']
```

#### Year Built Analysis

```python
# To retrieve the unique values in the yr_built column
unique_yr = df['yr_built'].unique()
print(sorted(unique_yr))
```

```python
# Exploring the trends in pricing and square footage over time using line plot
# Segment the data by year built
yearly_data = df.groupby('yr_built').agg({'price': 'mean', 'sqft_living': 'mean'})

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot trends in pricing over time
axs[0].plot(yearly_data.index, yearly_data['price'], color='blue', label='Average Price')
axs[0].set_xlabel('Year Built')
axs[0].set_ylabel('Average Price')
axs[0].set_title('Trends in Pricing Over Time')
axs[0].legend()

# Plot trends in average square footage over time
axs[1].plot(yearly_data.index, yearly_data['sqft_living'], color='green', label='Average Sqft Living')
axs[1].set_xlabel('Year Built')
axs[1].set_ylabel('Average Sqft Living')
axs[1].set_title('Trends in Average Square Footage Over Time')
axs[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

- Houses built in recent years have larger square footage of living area, i.e. there has been an increase over the years
